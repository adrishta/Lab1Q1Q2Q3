```python
import math
def distance(a,b):
   dis=0
   if len(a)==len(b):
    for i in range(len(a)):
      dis = dis+(a[i]-b[i]**2)
   return math.sqrt(dis)


v1=[3,6,7]
v2=[8,9,2]
distance(v1,v2)
```

```python
def f(x):
  return 1/(1+math.exp(-x))

def linspace(lower,upper,n):
  L=[]
  h=(upper-lower)/n
  L.append(lower)
  for i in range(n-1):
    L.append(L[i]+h)
  L.append(upper)

  return L


Listx=linspace(-10,10,500)
Listy=[]
for i in range(len(Listx)):
  Listy.append(f(Listx[i]))

print(Listx)
print(Listy)

import matplotlib.pyplot as plt
plt.plot(Listx,Listy)
plt.grid()
plt.show()
```
